### STEP2: NER PARSER 
* REMOVE TWEETS WITHOUT ENTITIES (PERSON, ORGANIZATION, LOCATION)
* SAVE TWEETS WITH ENTITIES INTO NEW COLLECTIONS
* DUE TO THE LARGE AMOUNT OF DATA, THIS PROCESS APLLIES MUITI PROCESSING FOR DOING THE NER PARSER

```python
def process_data(item):
    client = pymongo.MongoClient(host='localhost', port=27017)
    db_twitter=client.TwitterTP
    print(item)
    collection_twitter = db_twitter.Twitter1210_REMOVENOISE_NER

    doc=eng_tagger.tag(item['text'].split())
    #print(doc)
    orgstr = ""
    locstr = ""
    pstr = ""
    for i in range(len(doc)):
        if 'PERSON' in doc[i][1]:
            per =doc[i][0]
            pstr = pstr + "," + per
        #tag = 0
        else:
            if 'ORGANIZATION' in doc[i][1]:
                org =doc[i][0]
                orgstr = orgstr + "," + org
            else:
                if 'LOCATION' in doc[i][1]:
                    loc =doc[i][0]
                    locstr = locstr + "," + loc

    if (orgstr is not "") or (locstr is not "") or (pstr is not ""):
        print(item)
        id=item['id']
        created_at=item['created_at']
        entities=item['entities']
        geo=item['geo']
        place=item['place']
        coordinates=item['coordinates']
        timestamp_ms=item['timestamp_ms']
        result=collection_twitter.insert_one({"id":id,"ORGANIZATION": orgstr,"LOCATION": locstr,"PERSON": pstr,"ENTITIES":orgstr+locstr+pstr,
                                  "created_at":created_at,"entities":entities,"geo":geo,"coordinates":coordinates,
                                 "place":place,"timestamp_ms":timestamp_ms})
    #db_twitter.Twitter1210_REMOVENOISE.deleteOne({"id":id})
    
    client.close()
```

```python
import multiprocessing
import nltk
import numpy as np
import tweepy
from tweepy import Stream
from tweepy import OAuthHandler
from tweepy.streaming import StreamListener
import time
import json
import pymongo
from pymongo import MongoClient,ASCENDING, DESCENDING
import json
from pymongo import InsertOne, DeleteOne, ReplaceOne
from nltk.tag import StanfordNERTagger
eng_tagger = StanfordNERTagger(model_filename=r'C:\Users\User\Desktop\StanfordNLP\jars\stanford-ner-2018-10-16\classifiers\english.all.3class.distsim.crf.ser.gz',path_to_jar=r'C:\Users\User\Desktop\StanfordNLP\jars\stanford-ner-2018-10-16\stanford-ner.jar')
import queue
import threading
import datetime
exitFlag = 0

if __name__ == "__main__":
    client = pymongo.MongoClient(host='localhost', port=27017)
    db_twitter=client.TwitterTP

    starttime = datetime.datetime.now()
    
    manager = multiprocessing.Manager()
    queue = manager.Queue()
    
    for word in db_twitter.Twitter1210_REMOVENOISE.find().limit(100):
        #printj(word)
        queue.put(word)
    print("finished filling queue!!!")
    
    pool = multiprocessing.Pool(processes = 10)
    while not queue.empty():
        item = queue.get()
        #print(item)
        pool.apply_async(process_data, (item, ))

    print("Mark~ Mark~ Mark~~~~~~~~")

    pool.close()
    pool.join() 
#     queue.close()
#     queue.join_thread()
    print("Sub-process(es) done.")
    
    endtime = datetime.datetime.now()
    print("Time used:",endtime - starttime)
```

####  REMOYVE THE DUPLICATE ENTITIES AND TWEETS' ID IN GROUPS THEN SORTED THEM

```python
from collections import OrderedDict
for i in db_twitter.Twitter_1203_REMOVENOISE_NER_entities_ingroupG.find():
    ENTITY_N=''
    UID_N=''
    ENTITY_N=list(OrderedDict.fromkeys(i["ENTITY"]))
    UID_N=list(OrderedDict.fromkeys(i["UNIQUEIDS"]))
    id=i["_id"]
    myquery = {"_id": id}
    newvalues = {"$set": {"ENTITYUNI":sorted(ENTITY_N),"UNIQUEIDSUNI":sorted(UID_N)}}
    result=db_twitter.Twitter_1203_REMOVENOISE_NER_entities_ingroupG.update_one(myquery, newvalues)
```

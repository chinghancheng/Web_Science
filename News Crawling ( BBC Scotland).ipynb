# Def (content)

```python
def news_content(news_url):
    c=[]
    c1=[]
    res = requests.get(news_url, headers = headers)
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text, 'lxml')

    for a in soup.select('.story-body__inner'):
        if (len(a.select('p'))>5):
            for i in range(5):
                content=a.select('p')[i].text
                print(content)
                c.append(content)
            return(c)
        else:
            for j in range(len(a.select('p'))):
                content=a.select('p')[j].text
                print(content)
                c1.append(content)
            return(c1)
```

# Def (Time)

```python
from datetime import datetime
def time_transfer(timeSS):
    from datetime import datetime
    ts = int(timeSS)

    # if you encounter a "year is out of range" error the timestamp
    # may be in milliseconds, try `ts /= 1000` in that case
    return(datetime.utcfromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S'))
```

```python
def timeS(url):
    res = requests.get(url, headers = headers)
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text, 'lxml')
    article_time=[]
    for news_time in soup.select('.mini-info-list__item'):
        if(len(news_time.select('div'))>0):  
            timeS=news_time.select('div')[0]
            print(time_transfer(timeS['data-seconds']))
            article_time.append(time_transfer(timeS['data-seconds']))
        else:
            print('None')
        return(article_time)
    
```

```python
def News(datasetname):
    article_url=[]
    article_title=[]
    article_time=[]
    article_content=[]
    client = pymongo.MongoClient(host='localhost', port=27017)
    db_News=client.News
    collection_News= db_News[datasetname]
    for i in range(len(soup.select('.title-link'))):
        a=soup.select('.title-link')[i]
        url=a['href']
        title=a.text
        if 'http://' in url:
            article_url.append(url)
            article_time.append(timeS(url))
            article_content.append(news_content(url))
        elif 'https://' in url:
            article_url.append(url)
            article_time.append(timeS(url))
            article_content.append(news_content(url))
        else:
            url='https://www.bbc.co.uk'+url
            article_url.append(url)
            article_time.append(timeS(url))
            article_content.append(news_content(url))
        article_title.append(title)
    for i in range(len(article_content)):
        d = {'article_time' :article_time[i],'article_title': article_title[i], 'article_content' :article_content[i], 'article_url' : article_url[i]}
        result=collection_News.insert_one(d)
        print(result)
```
